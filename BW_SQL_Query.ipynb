```python
no_of_quotes='select * from quotes limit 5'
pd.read_sql(no_of_quotes,engine)
```

##### Find the number of quotes by each author


```python
no_of_quotes='''SELECT 
    author, COUNT(*) AS quote_count
FROM
    quotes
GROUP BY author
ORDER BY quote_count DESC'''
pd.read_sql(no_of_quotes,engine)
```

##### List the top 5 most common tags

```python
top_common_tags='''select tags,count(*)  as tag_count
from quotes
group by tags
order by count(*) desc limit 5'''
pd.read_sql(top_common_tags,engine)
```

##### Find authors who have more than 5 quotes.


```python
quote='''select author ,count(*) as quotes_count
from quotes
group by author
having  count(*)>5'''
pd.read_sql(quote,engine)
```

##### Retrieve the longest quote and its author.**


```python

longest_quote='''select *,length(quote) as quote_len
from quotes
order by length(quote) desc
limit 1'''
pd.read_sql(longest_quote,engine)
```
