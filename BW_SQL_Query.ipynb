# - 1. Find the number of quotes by each author.**

```python
x='select * from quotes limit 5'
pd.read_sql(x,engine)
```

```python
x='''select tag_name,count(*)  as tag_count
from quotes
group by tag_name
order by count(tag_name) desc limit 5'''
pd.read_sql(x,engine)
```

```python
#Find authors who have more than 5 quotes.
x='''select author ,count(*) as quotes_count
from quotes
group by author
having  count(*)>5'''
pd.read_sql(x,engine)
```

```python
# Retrieve the longest quote and its author.**

x='''select *,length(quote) as quote_len
from quotes
order by length(quote) desc
limit 1'''
pd.read_sql(x,engine)
```
